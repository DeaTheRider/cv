```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import imageio
```

```python
img = imageio.imread('cat-color.png')
img = img.astype(np.float32) / 255.
```

```python
from skimage.color import rgb2hsv, hsv2rgb
```

```python
img_hsv = rgb2hsv(img)
```

```python
def to_hsv(img):
    out = np.zeros(img.shape)
    v = img.max(axis=2)
    temp = img.min(axis=2)
    s = np.zeros(v.shape)
    h = np.zeros(v.shape)
    for i in range(v.shape[0]):
        for j in range(v.shape[1]):
            if v[i,j] == 0:
                s[i,j] = 0
            else:
                s[i,j] = (v[i,j]-temp[i,j])/v[i,j]
            if v[i,j] == img[i,j,0] and img[i,j,1] >= img[i,j,2]:
                h[i,j] = 60*(img[i,j,1] - img[i,j,2])/(v[i,j] - temp[i,j]) + 0
            elif v[i,j] == img[i,j,0] and img[i,j,1] < img[i,j,2]:
                h[i,j] = 60*(img[i,j,1] - img[i,j,2])/(v[i,j] - temp[i,j]) + 360
            elif v[i,j] == img[i,j,1]:
                h[i,j] = 60*(img[i,j,2] - img[i,j,0])/(v[i,j] - temp[i,j]) + 120
            elif v[i,j] == img[i,j,2]:
                h[i,j] = 60*(img[i,j,0] - img[i,j,1])/(v[i,j] - temp[i,j]) + 240
    out[..., 0] = h / 360
    out[..., 1] = s
    out[..., 2] = v 
    
            
            
                
    return out
```

```python
def to_rgb(img):
    h = np.array(img[:,:,0]*360, dtype=int)
    s = np.array(img[:,:,1]*100, dtype=int)
    v = np.array(img[:,:,2]*100, dtype=int)
    Hi = (h/60)%6
    Vmin = (100-s)*v//100
    a = (v-Vmin)*(h%60)//60
    Vinc = Vmin + a
    Vdec = v - a
    Hi = np.stack([Hi, Hi, Hi], axis=-1).astype(np.uint8) % 6
                    
    out = np.choose(Hi,[np.stack((v, Vinc, Vmin), axis=2),
                         np.stack((Vdec, v, Vmin), axis=2),
                         np.stack((Vmin, v, Vinc), axis=2),
                         np.stack((Vmin, Vdec, v), axis=2),
                         np.stack((Vinc, Vmin, v), axis=2),
                         np.stack((v, Vmin, Vdec), axis=2)])
    
    
            
    return out    
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(to_hsv(img))
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(to_rgb(img_new))
```
